Task 4: Image to Pencil Sketch with Python:

```python
#importing the necessary libraries
import cv2
from matplotlib import pyplot as pyplot
```

```python
#read the image
image= cv2.imread("selenagomez.jpg")
```

```python
#path of the image
path = r'C:\Users\YASH\OneDrive\Desktop\selenagomez.jpg'
```

```python
##Reading an image in default mode
img=cv2.imread(path,0)
```

```python
#Window name in which image is displayed
window_name='img'
```

```python
## Using cv2.imshow() method 
# Displaying the image 
cv2.imshow(window_name,img)
cv2.waitKey(0)
```

```python
#read image in RGB format
from matplotlib import pyplot as plt
%matplotlib inline
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
#Now after reading the image, we will create a new image by converting the original image to greyscale:
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray_img
```

```python
##show the gray image
plt.imshow(gray_img)
```

```python
#Now the next step is to invert the new grayscale image:
inverted_img = 255-gray_img
plt.imshow(inverted_img)
```

```python
#Now the next step in the process is to blur the image by using the Gaussian Function in OpenCV:
blurred = cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blurred)
```

```python
#converting to pencil sketch
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_img,inverted_blurred,scale=268.0)
plt.imshow(pencil_sketch)
cv2.waitKey(0)
```

```python
##And finally, if you want to have a look at both the original image and the pencil sketch then you can use the following commands:
cv2.imshow("img", img)
cv2.imshow("pencil_sketch", pencil_sketch)
cv2.waitKey(0)
```
